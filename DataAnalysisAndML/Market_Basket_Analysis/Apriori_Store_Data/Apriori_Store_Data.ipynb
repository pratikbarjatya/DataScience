```python
import pandas as pd
from apyori import apriori
```

```python
df_store=pd.read_csv("store_data.csv",header=None)
df_store.fillna("Oracle",inplace=True)
```

```python
df_store
```

```python
liste = df_store.astype(str).values.tolist()
```

```python
liste
```

```python
data_len=df_store.shape[0]
col_num=df_store.shape[1]
print(data_len,col_num)
```

```python
final_list=[]
for i in range(0,data_len):
    eachRecList=[]
    for j in range(0,20):
#         print(i,j,df_store.values[i,j])
        if((~pd.isna(df_store.values[i,j])) & (df_store.values[i,j]!="Oracle")):
              eachRecList.append(str(df_store.values[i,j]))
    final_list.append(eachRecList)
    
print(final_list[0])    
        
```

```python
type(final_list[0])
```

```python
association_rules=apriori(final_list,min_support=.005,min_confidence=.20,min_lift=3,min_length=2)
```

```python
listofRules=list(association_rules)
```

```python
rulesList_len=len(listofRules)
rulesList_len
```

```python
def decode_association_rule(listofRules):
    final_list=[]
    for eachRec in listofRules:
        final_dict={}
        print(eachRec[2])
        ord_stats_list=eachRec[2]
        final_dict["Items"]=",".join(eachRec[0])
        final_dict["Support"]=round(eachRec[1],3)
        final_dict["Item Base"]=",".join(ord_stats_list[0][0])
        final_dict["Item add"]=",".join(ord_stats_list[0][1])
        final_dict["Confidence"]=round(ord_stats_list[0][2],3)
        final_dict["Lift"]=round(ord_stats_list[0][3],3)
        final_list.append(final_dict)
    return final_list
                                    

```

```python
association_rule=decode_association_rule(listofRules)
association_rule
```

```python
df=pd.DataFrame(association_rule,columns=["Items","Support","Item Base","Item add","Confidence","Lift"])
```
