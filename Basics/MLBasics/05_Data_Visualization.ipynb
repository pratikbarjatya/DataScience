```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
year = [2015, 2016, 2017, 2018, 2019]
sales = [3456,4445,3333,5456,5555]
profit = [345,555,768,666,778]
```

```python
plt.plot(year,sales)
plt.show()
```

```python
plt.plot(year, sales, 'red')
plt.plot(year, profit, 'blue')
plt.title('Sales and Profit vs Years')
plt.xlabel('Years')
plt.ylabel('Sales and Profit')
plt.show()
```

```python
plt.axes([0.05, 0.05,0.9,0.9])
plt.plot(year, sales, 'red')
plt.title('Year vs Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.axes([1.1, 0.05,0.9,0.9])
plt.plot(year, profit, 'blue')
plt.title('Year vs Profit')
plt.xlabel('Year')
plt.ylabel('Profit')
```

```python
plt.subplot(2,1,1)
plt.plot(year, sales, 'red')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.title('Sales by Years')

plt.subplot(2,1,2)
plt.plot(year, profit, 'blue')
plt.xlabel('Years')
plt.ylabel('Profit')
plt.title('Profit by Years')
plt.tight_layout()
```

```python
plt.plot(year, sales, 'red')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.title('Sales by Years')
plt.xlim((2017,2018))
plt.ylim((3000,4000))
```

```python
plt.plot(year, sales, 'red')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.title('Sales by Years')
plt.axis((2017,2018,3000,4000))
```

```python
plt.scatter(y = profit, x=sales, marker='o', label='Sales vs Profit', color='red')
plt.legend()
plt.annotate('Customer1', xy=(4445,555))
plt.style.use('seaborn-darkgrid')
```
