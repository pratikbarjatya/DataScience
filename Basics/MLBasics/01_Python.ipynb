```python
flight_data = pd.read_csv('flight_data.csv')
```

```python
flight_data
```

```python
(flight_data >>
 select (X.origin, X.dest, X.hour))
```

```python
(flight_data>>
drop(X.year, X.month, X.day))
```

```python
(flight_data>>
select(~X.hour, ~X.minute))
```

```python
(flight_data>>
mask(X.month==1, X.day==1, X.origin=='JFK', X.hour>10))
```

```python
(flight_data>>
arrange(X.distance, X.hour,ascending=False))
```

```python
(flight_data>>
mask(X.origin=='EWR')>>
mutate(new_distance=X.distance/1000,
      carrier_origin = X.carrier + X.origin))
```

```python
(flight_data>>
group_by(X.origin))
```

```python
(flight_data>>
group_by(X.origin)>>
summarise(mean_disctance=X.distance.min()))
```

```python
(flight_data>>
 mask(X.hour>10)>>
mutate(speed = X.distance/(X.air_time*60))>>
group_by(X.origin)>>
summarise(mean_speed=X.speed.mean())>>
arrange(X.mean_speed, ascending=False))
```
